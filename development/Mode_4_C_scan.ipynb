When mode 4 is active ABI captures full disk images every 5 minutes but no other scans. We should extact missing CONUS scan data from this when it's running

```python
import numpy as np
from numpy import ma
import pandas as pd
import xarray as xr
from datetime import datetime, timedelta
from dateutil.parser import parse as parse_date
from tobac_flow import io
```

```python
# Load files
goes_data_path = '../data/GOES16'
date = datetime(2018,6,19,17)

abiC_files = sorted(io.find_abi_files(date, satellite=16, product='MCMIP', view='C', mode=3, 
                               save_dir=goes_data_path, 
                               replicate_path=True, check_download=True, 
                               n_attempts=1, download_missing=True))

abiF_files = sorted(io.find_abi_files(date, satellite=16, product='MCMIP', view='F', mode=3, 
                               save_dir=goes_data_path, 
                               replicate_path=True, check_download=True, 
                               n_attempts=1, download_missing=True))

print("CONUS files:", len(abiC_files))
print("Full disk files:", len(abiF_files))
      
```

```python
abiC_ds = xr.open_dataset(abiC_files[0])
abiF_ds = xr.open_dataset(abiF_files[0])
```
